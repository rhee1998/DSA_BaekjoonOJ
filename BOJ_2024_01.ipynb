```python
# 1711 직각삼각형 [G5]
import sys
input = sys.stdin.readline

# ======================
def gcd(x, y):
    x, y = abs(x), abs(y)
    if y > x: return gcd(y, x)
    if y == 0: return x
    return gcd(y, x % y)

def reduce(x, y):
    g = gcd(x, y)
    return x // g, y // g
# ======================

N, P = int(input()), []
result = 0

for _ in range(N):
    x, y = map(int, input().split())
    P.append([x, y])

for x0, y0 in P:
    slopes = {}
    for x, y in P:
        x_rel, y_rel = x - x0, y - y0
        if x_rel == 0 and y_rel == 0:
            continue
        x_rel, y_rel = reduce(x_rel, y_rel)
        
        try:
            slopes[x_rel, y_rel] += 1
        except:
            slopes[x_rel, y_rel] = 1

    for x, y in slopes.keys():
        try:
            result += slopes[x, y] * slopes[-y, x]
        except:
            continue

print(result)
```

```python
# 21736 헌내기는 친구가 필요해 [S2]
import sys
from collections import deque

N, M = map(int, sys.stdin.readline().split())
campus = []
visited = [[0 for _ in range(M)] for _ in range(N)]
q = deque([])

for n in range(N):
    row = [x for x in sys.stdin.readline().rstrip()]
    campus.append(row)
    for m in range(M):
        if row[m] == 'I':
            q.append((n, m))
            visited[n][m] = 1

result = 0
while q:
    n, m = q.popleft()
    for (x, y) in [(n + 1, m), (n - 1, m), (n, m + 1), (n, m - 1)]:
        if x < 0 or x >= N: continue
        if y < 0 or y >= M: continue

        if visited[x][y] == 1: continue
        if campus[x][y] == 'X': continue
        if campus[x][y] == 'P': result += 1
    
        visited[x][y] = 1
        q.append((x, y))

if result == 0: print('TT')
else: print(result)
```

```python
# 14940 쉬운 최단거리 [S1]
import sys
from collections import deque

N, M = map(int, sys.stdin.readline().split())
world = []
visited = [[0 for _ in range(M)] for _ in range(N)]
dists = [[-1 for _ in range(M)] for _ in range(N)]
q = deque([])

for n in range(N):
    row = [int(x) for x in sys.stdin.readline().split()]
    world.append(row)
    for m in range(M):
        if row[m] == 2:
            q.append((n, m))
            visited[n][m] = 1
            dists[n][m] = 0

while q:
    n, m = q.popleft()
    for (x, y) in [(n + 1, m), (n - 1, m), (n, m + 1), (n, m - 1)]:
        if x < 0 or x >= N: continue
        if y < 0 or y >= M: continue

        if visited[x][y] == 1: continue
        if world[x][y] == 0: continue
    
        dists[x][y] = dists[n][m] + 1
        visited[x][y] = 1
        q.append((x, y))

for n in range(N):
    s = ''
    for m in range(M):
        if world[n][m] == 0: dists[n][m] = 0
        s = f'{s}{dists[n][m]} '
    sys.stdout.write(s + '\n')
```

```python
# 1865 웜홀 [G3]
import sys
input = sys.stdin.readline

def BellmanFord(start, N, graph, dist, wdict, INF=10**30):
    res = {v : INF for v in graph.keys()}
    res[start] = 0
    neg_cycle = False

    for i in range(N):
        for s, e in dist.keys():
            try:
                if res[s] != INF and res[e] > res[s] + dist[s, e]:
                    res[e] = res[s] + dist[s, e]
                    
                    # if res is updated in the last loop, negative cycle exists
                    if i == N - 1: neg_cycle = True
            except:
                continue

    for w in wdict.keys():
        if res[w] < INF: wdict[w] = 1

    return neg_cycle, wdict

for _ in range(int(input())):
    N, M, W = map(int, input().split())
    G, D = {v : [] for v in range(N)}, {}
    for _ in range(M):
        s, e, t = map(int, input().split())
        s, e = s - 1, e - 1
        G[s].append(e)
        G[e].append(s)

        try:
            D[s, e] = min([D[s, e], t])
            D[e, s] = min([D[e, s], t])
        except:
            D[s, e] = t
            D[e, s] = t

    worms = []
    for _ in range(W):
        s, e, t = map(int, input().split())
        s, e, t = s - 1, e - 1, -t
        G[s].append(e)
        worms.append(s)

        try:
            D[s, e] = min([D[s, e], t])
        except:
            D[s, e] = t

    worms = list(set(worms))
    wdict = {w : 0 for w in worms}

    ans = 'NO'
    for w in worms:
        if wdict[w] == 1: continue
        neg_cycle, wdict = BellmanFord(w, N, G, D, wdict)
        if neg_cycle:
            ans = 'YES'
            break
    print(ans)
```

```python
# 23048 자연수 색칠하기 [G5]
import sys
N = int(input())

primes = [1 for _ in range(N + 1)]
for p in range(2, N):
    if p ** 2 > N: break
    if primes[p] == 0: continue
    for x in range(2 * p, N + 1, p): primes[x] = 0

ans, K = [1, 1] + [0] * (N - 1), 1
for i in range(2, N + 1):
    if primes[i] == 1:
        K += 1
        for x in range(i, N + 1, i): ans[x] = K

sys.stdout.write(f'{K}\n')
for x in ans[1:]:
    sys.stdout.write(f'{x} ')
```

```python
# 1197 최소 스패닝 트리 [G4] - Prim Algorithm
import sys, heapq
input = sys.stdin.readline

V, E = map(int, input().split())
G = {v : [] for v in range(V)}

for _ in range(E):
    a, b, c = map(int, input().split())
    a, b = a - 1, b - 1
    G[a].append((b, c))
    G[b].append((a, c))

visited = [False for _ in range(V)]

# Initialize Priority Queue
que = []
visited[0] = True
for next, cost in G[0]:
    heapq.heappush(que, (cost, next))

# Iterate
res = 0
while que:
    cost, curr = heapq.heappop(que)
    
    if visited[curr] == True: continue
    
    visited[curr] = True
    res += cost

    for next, cost in G[curr]:
        if visited[next] == True: continue
        heapq.heappush(que, (cost, next))

print(res)
```

```python
# 1197 최소 스패닝 트리 [G4] - Kruskal's Algorithm
import sys, heapq
input = sys.stdin.readline

V, E = map(int, input().split())
edges = []

for _ in range(E):
    a, b, c = map(int, input().split())
    a, b = a - 1, b - 1
    heapq.heappush(edges, (c, a, b))

# Union-Find Algorithm
parent = [v for v in range(V)]
rank = [0 for _ in range(V)]

def FindRoot(x):
    while x != parent[x]:
        x = parent[x]
    return x

def Union(x, y):
    px, py = FindRoot(x), FindRoot(y)
    
    if rank[px] < rank[py]: parent[px] = py
    else: parent[py] = px

    if rank[px] == rank[py]: rank[px] += 1
    return

# Kruskal's Algorithm
def KruskalMST(edges):
    mst = []
    while edges:
        cost, x, y = heapq.heappop(edges)
        if FindRoot(x) == FindRoot(y):
            continue
        else:
            Union(x, y)
            mst.append((cost, x, y))

    return mst

mst = KruskalMST(edges)
res = 0
for cost, _, _ in mst:
    res += cost

print(res)
```

```python
# 2887 행성터널 [P5]
import sys, heapq

V = int(input())
P = []
for v in range(V):
    x, y, z = map(int, input().split())
    P.append((v, x, y, z))

def Compare(x, y, priority):
    for p in priority:
        if x[p] < y[p]: return True
        if x[p] > y[p]: return False
    return True

def MergeSort(A, priority):
    L = len(A)
    if L <= 1: return A

    F, B = MergeSort(A[:L//2], priority), MergeSort(A[L//2:], priority)
    fidx, bidx, R = 0, 0, []

    while fidx < L//2 and bidx < L - L//2:
        if Compare(F[fidx], B[bidx], priority):
            R.append(F[fidx]); fidx += 1
        else:
            R.append(B[bidx]); bidx += 1

    R += F[fidx:] + B[bidx:]
    return R

X = MergeSort(P, [1])
Y = MergeSort(P, [2])
Z = MergeSort(P, [3])

G = {v : [] for v in range(V)}
for i in range(V - 1):
    p, q = X[i][0], X[i+1][0]
    G[p].append((q, P[q][1] - P[p][1]))
    G[q].append((p, P[q][1] - P[p][1]))

    p, q = Y[i][0], Y[i+1][0]
    G[p].append((q, P[q][2] - P[p][2]))
    G[q].append((p, P[q][2] - P[p][2]))

    p, q = Z[i][0], Z[i+1][0]
    G[p].append((q, P[q][3] - P[p][3]))
    G[q].append((p, P[q][3] - P[p][3]))

# Initialize Priority Queue
visited = [False for _ in range(V)]
visited[0] = True

que = []
for next, cost in G[0]:
    heapq.heappush(que, (cost, next))

# Iterate
res = 0
while que:
    cost, curr = heapq.heappop(que)
    
    if visited[curr] == True: continue
    
    visited[curr] = True
    res += cost

    for next, cost in G[curr]:
        if visited[next] == True: continue
        heapq.heappush(que, (cost, next))

print(res)
```
